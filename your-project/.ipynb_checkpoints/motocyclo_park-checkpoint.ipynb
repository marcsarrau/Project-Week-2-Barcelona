### QUESTION 1: HOW DID THE MOTORCYCLE PARK EVOLVE BETWEEN 2007 AND 2018?

```python
import pandas as pd
import numpy as np
```

#### Motorcycle Potential Rider Population

```python
population = pd.read_csv('../questions/barcelona_datasets/poblacio/padro/2007_padro_edat_any_a_any_per_sexe.csv')
population = population[population['Edat_any_a_any'].isin(['15 anys', '16 anys', '17 anys', '18 anys', '19 anys', '20 anys', '21 anys',
                                                           '22 anys', '23 anys', '24 anys', '25 anys', '26 anys', '27 anys', '28 anys', 
                                                           '29 anys', '30 anys', '31 anys', '32 anys', '33 anys', '34 anys', '35 anys',
                                                           '36 anys', '37 anys', '38 anys', '39 anys', '40 anys', '41 anys', '42 anys',
                                                           '43 anys', '44 anys', '45 anys', '46 anys', '47 anys', '48 anys', '49 anys',
                                                           '50 anys', '51 anys', '52 anys', '53 anys', '54 anys', '55 anys', '56 anys',
                                                           '57 anys', '58 anys', '59 anys', '60 anys', '61 anys', '62 anys', '63 anys',
                                                           '64 anys', '65 anys', '66 anys', '67 anys', '68 anys', '69 anys', '70 anys',
                                                           '71 anys', '72 anys', '73 anys', '74 anys', '75 anys'])]
                                                                                                                       
population
```

```python
population_ct = pd.crosstab(index=population['Nom_Districte'], columns=['Nombre'],
                  values=population['Nombre'], aggfunc='sum')
population_ct
```

```python
sum_population_ct = population_ct['Nombre'].sum()
sum_population_ct
```

#### Vehicle Total Park

```python
total_park = pd.read_csv('../questions/barcelona_datasets/vehicles/car_park_tipology/2007_tipologia_parc_vehicles.csv')
total_park = total_park[total_park.Nom_Districte != "No consta"]
total_park
```

```python
sum_total_park = total_park['Nombre_vehicles'].sum()
sum_total_park
```

#### Motorcycle Total Park

```python
motocyclo_park = total_park[total_park['Tipologia_parc'].isin(['Motos', 'Ciclomotors'])]
motocyclo_park
```

```python
sum_motocyclo_park = motocyclo_park['Nombre_vehicles'].sum()
sum_motocyclo_park
```

#### Absolut Values of Motorcycle Total Park and Potential Riders

```python
motocyclo_park_abs = pd.crosstab(index=motocyclo_park['Nom_Districte'], columns=['Tipologia_parc'], values=motocyclo_park['Nombre_vehicles'], aggfunc=['sum'])
motocyclo_park_abs = motocyclo_park_abs.merge(right=population_ct, how='left', on='Nom_Districte')
motocyclo_park_abs.columns = ['Motos & Cyclomotos', 'Potential Riders']
motocyclo_park_abs.index.names = ['2007']
motocyclo_park_abs
```

```python
motocyclo_park_abs.to_csv('motocyclo_park_abs_2007.csv')
```

#### % Ratio of Motorcycles per Potential Riders

```python
columns = motocyclo_park_abs[:-1]
columns

for col in columns:
    motocyclo_park_abs[col] = motocyclo_park_abs[col]/motocyclo_park_abs['Potential Riders']*100
    motocyclo_park_rat = motocyclo_park_abs
motocyclo_park_rat.index.names = ['2007']
motocyclo_park_rat
```

```python
motocyclo_park_rat.to_csv('motocyclo_park_rat_2007.csv')
```
