```python
import pandas as pd
import numpy as np
```

```python
population = pd.read_csv('../questions/barcelona_datasets/poblacio/padro/2007_padro_edat_any_a_any_per_sexe.csv')
population = population[population['Edat_any_a_any'].isin(['15 anys', '16 anys', '17 anys', '18 anys', '19 anys', '20 anys', '21 anys',
                                                           '22 anys', '23 anys', '24 anys', '25 anys', '26 anys', '27 anys', '28 anys', 
                                                           '29 anys', '30 anys', '31 anys', '32 anys', '33 anys', '34 anys', '35 anys',
                                                           '36 anys', '37 anys', '38 anys', '39 anys', '40 anys', '41 anys', '42 anys',
                                                           '43 anys', '44 anys', '45 anys', '46 anys', '47 anys', '48 anys', '49 anys',
                                                           '50 anys', '51 anys', '52 anys', '53 anys', '54 anys', '55 anys', '56 anys',
                                                           '57 anys', '58 anys', '59 anys', '60 anys', '61 anys', '62 anys', '63 anys',
                                                           '64 anys', '65 anys', '66 anys', '67 anys', '68 anys', '69 anys', '70 anys',
                                                           '71 anys', '72 anys', '73 anys', '74 anys', '75 anys'])]
                                                                                                                       
population
```

```python
population_ct = pd.crosstab(index=population['Nom_Districte'], columns=['Nombre'],
                  values=population['Nombre'], aggfunc='sum')
population_ct
```

```python
cilinders = pd.read_csv('../questions/barcelona_datasets/vehicles/motorcycles_cilinders/2007_cilindrada_motos.csv')
cilinders["Cilindrada_motos"].replace({"Fins a 75 cc" : "1. Below 75 cc", 
                                       "de 75 a 125 cc" : "2. From 75 to 125 cc",
                                       "de 126 a 200 cc" : "3. From 126 to 200 cc",
                                       "de 201 a 250 cc" : "4. From 201 to 250 cc",
                                       "de 251 a 300 cc" : "5. From 251 to 300 cc",
                                       "de 301 a 500 cc" : "6. From 301 to 500 cc",
                                       "de 501 a 750 cc" : "7. From 501 to 750 cc",
                                       "Més de 750 cc" : "8. More than 750 cc"
                                       }, inplace=True)
cilinders = cilinders[cilinders.Nom_Districte != "No consta"]
cilinders = cilinders[cilinders.Cilindrada_motos != "No consta"]
cilinders
```

```python
cilinders_ct = pd.crosstab(index=cilinders['Nom_Districte'], columns=cilinders['Cilindrada_motos'], values=cilinders['Nombre_motos'], aggfunc=['sum'])
cilinders_ct = cilinders_ct.merge(right=population_ct, how='left', on='Nom_Districte')
cilinders_ct.columns = ['Below 75 cc', 'From 75 to 125 cc', 'From 126 to 200 cc', 'From 201 to 250 cc',
                              'From 251 to 300 cc', 'From 301 to 500 cc', 'From 501 to 750 cc', 'More than 750 cc', 'Potential Riders']
cilinders_ct.index.names = ['2007']
cilinders_ct
```

```python
#cilinders_ct.to_csv('cilinders_abs_2007.csv')
```

```python
columns = cilinders_ct.columns[:]
columns

for col in columns:
    cilinders_ct[col] = cilinders_ct[col]/cilinders_ct['Potential Riders']*100
cilinders_ct
```
